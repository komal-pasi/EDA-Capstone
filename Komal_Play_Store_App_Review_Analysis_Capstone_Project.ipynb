#**What is EDA?**
#**EDA means  Exploratory Data Analysis**
Exploratory Data Analysis, or EDA, is an important step in any Data Analysis or Data Science project. EDA is the process of investigating the dataset to discover patterns, and anomalies (outliers), and form hypotheses based on our understanding of the dataset.

EDA involves generating summary statistics for numerical data in the dataset and creating various graphical representations to understand the data better.

Now, we will perform some data cleaning operations on the dataset to get clean data for EDA.

Here we have removed '+' & ',' symbols from 'Installs' columns.

```python
ps_dataset['Installs']=ps_dataset['Installs'].apply(lambda i:i.replace(',' ,'') if ','  in str(i) else i)
ps_dataset['Installs']=ps_dataset['Installs'].apply(lambda i:i.replace('+','') if '+' in str(i) else i)
ps_dataset
```

So now, let's define a function that will give us more detailed information about the given dataset such as total null values in each column, total no. of unique values in each column.
There's also a advantage of defining a function i.e we can it in future codes too...


```python
def detailed_info(dataset):
    f1 = pd.DataFrame(index=dataset.columns)
    f1['count_of_nulls'] = dataset.isnull().sum()
    f1['count_of_uniques'] = dataset.nunique()
    f1['data_types'] = dataset.dtypes
    return f1
```

```python
detailed_info(ps_reviews)
```

Now, let's have look on data types of the columns. There are some columns with 'object' data types which should have been int/float.
So let's check the data of these columns and convert them to correct data types 

```python
#Converted data type of'Reviews' column to int using .astype()
ps_dataset['Reviews'] = ps_dataset['Reviews'].astype(int)
```

```python
#See whether its converted or not
detailed_info(ps_dataset)
```

Now, let's replace al the dollor sign from 'Price' column and convert it's datatype into float.

```python
#Replacing $ using lambda function. And then converted it into Float datatype.
ps_dataset['Price'] = ps_dataset['Price'].apply(lambda x: x.replace('$','') if '$' in str(x) else x)
ps_dataset['Price'] = ps_dataset['Price'].apply(lambda x: float(x))
```

Let's check if the datatype of 'Price' column is converted to float.

```python
detailed_info(ps_dataset)
```

Now here will perform following operations in order to clean 'Size column'


**Operations:**

1.   Replacing String 'Varies with Device' with Nan
2.   Removing comman(,)
3.   Replacing letter 'M'
4.   Convert kilobyte to megabyte and replace letter 'k'
5.   Lastly converting datatype into float.


```python
ps_dataset['Size'] = ps_dataset['Size'].apply(lambda x: str(x).replace('Varies with device', 'NaN') if 'Varies with device' in str(x) else x)
ps_dataset['Size'] = ps_dataset['Size'].apply(lambda x: str(x).replace('M','') if 'M' in str(x) else x)
ps_dataset['Size'] = ps_dataset['Size'].apply(lambda x: str(x).replace(',','') if 'M' in str(x) else x)
ps_dataset['Size'] = ps_dataset['Size'].apply(lambda x: float(str(x).replace('k',''))/1024 if 'k' in str(x) else x)
```

####**Last Updated:**
Now will convert the data type of this feature into pandas datetime.

```python
# convert the ' Last Updated' column to pandas dataframe
ps_dataset['Last Updated'] = ps_dataset['Last Updated'].apply(lambda x: pd.to_datetime(x))
```

Now, here We will start with Visualizing our data analysis.


Here will see that How is Rating effected by Price of app.

```python
sns.set_theme(style='whitegrid',rc={'figure.figsize': (10,12)},palette='Spectral')
sns.lmplot(y='Rating',x='Price',data=ps_dataset)
pt.grid(False)  #Remove Grid lines
```

Now will see whether there's any correlation between Ratings and Reviews.

```python
sns.set_theme(style='whitegrid',palette='Spectral',rc={'figure.figsize': (8,8)})
sns.lmplot(y='Rating',x='Reviews',data=ps_dataset)
pt.grid(False)
for spine in pt.gca().spines.values():
  spine.set_visible(False)
```

In above shown chart we can clearly see there's no correlation between these two columns.

Now we will see how many percentage of apps are paid and how many are free using pie chart.

```python
paid_free_app_df = ps_dataset.groupby('Type').agg({"App": "count"}).reset_index()
sns.set_theme(style='whitegrid',palette='muted',rc={'figure.figsize': (8,8)})
data = paid_free_app_df['App']
labels = paid_free_app_df['Type']
colors = ["#2E3192","#CC1313"]
explode=(0,0.05)
pt.pie(data, labels = labels, colors=colors, autopct='%.0f%%',explode=explode,textprops={'fontsize': 18,'color':"w"})
pt.title('Distribution among Free and Paid apps',size=15,loc='center')
pt.axis('equal')
pt.show()
```

Now, will see the ratio of updation of apps being getting by years.

```python
last_update_df = ps_dataset.copy()
last_update_df['Updating_Year'] = pd.DatetimeIndex(last_update_df['Last Updated']).year
```

```python
update_years = last_update_df.groupby('Updating_Year').agg({"App": "count"}).reset_index()
sns.lineplot(data=update_years, x="Updating_Year", y="App",color="#2E3192")
pt.title('Trend showing last Update',size=14)
pt.ylabel('Last Update Apps Count',size=12)
pt.xlabel('Years',size=12)
pt.grid(False)
for spine in pt.gca().spines.values():
  spine.set_visible(False)
```

Now, will see Whether the recently updated apps are more liked by the users or not.

```python
sns.set_theme(style="whitegrid",palette='Spectral',rc={'figure.figsize': (12,8)})
sns.regplot(data=last_update_df,x='Updating_Year',y='Rating',color="#2E3192")
pt.title('Ratings changing as per last Updates',size=18)
for spine in pt.gca().spines.values():
  spine.set_visible(False)
pt.grid(False)
```

In the above chart we can see that Apps which get updated frequently have more chances of geting more ratings.
